```python
doc_indexes = {}
with open('./urls.numerate.txt', 'r') as f:
    while f.readable():
        line = f.readline()
        
        if line.strip() == '':
            break
        
        idx, url = line.strip().split('\t')
        idx = int(idx)
        doc_indexes[url] = idx
```

```python
exceptions = []
for d in main_dirs[4:]:
    files = os.listdir('./content/' + d)
    for f_name in tqdm(files):
        gc.collect()
        
        f_id = int(f_name.split('.')[1])
        
        if d == '20170710' and f_id <= 4589:
            continue
        
        text = ''
        
        try:
            with open('./content/' + d + '/' + f_name, 'r') as file:
                text = file.read()
        except Exception:
            exceptions.append('./content/' + d + '/' + f_name)
            continue
        
        result_text = ''
        
        
        extractor = Extractor(extractor='ArticleExtractor', html=text)
        idx = doc_indexes[extractor.data.split('\n')[0]]
        
        try:
            with open('./texts/' + str(idx) + '.txt', 'w') as fout:
                fout.write(extractor.getText())
        except Exception:
            exceptions.append('./content/' + d + '/' + f_name)
            continue
```
